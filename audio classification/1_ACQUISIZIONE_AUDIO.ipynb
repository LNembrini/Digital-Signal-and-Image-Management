```python
import sounddevice as sd
from scipy.io.wavfile import write
import speech_recognition as sr
from gtts import gTTS
import os
import time
import playsound
import numpy as np
import librosa
```

```python
def audio(name, command):
    
    fs= 88200
    seconds= 3
    count= 0
    
    for i in range(50):
        fold= name+'_'+command
        if not os.path.exists(fold):
            os.mkdir(fold)
        
        print('Start recording ', str(i+1)+'/50')
        myrecording= sd.rec(int(seconds * fs), samplerate=fs, channels=1)
        sd.wait()  # Wait until recording is finished
        
        #noise
        noise1= np.random.randn(len(myrecording)).reshape(-1,1)
        noise= (myrecording + 0.005*noise1).astype('float32')
        
        #pitch
        data= myrecording.reshape(-1)
        pitch= librosa.effects.pitch_shift(data, 88200, -.8).reshape(-1,1)
        
        #speed
        data= myrecording.reshape(-1)
        speed= librosa.effects.time_stretch(data, 1.5).reshape(-1,1)
                
        records= [myrecording, noise, pitch, speed]
        
        for r in records:
            write(fold+'/'+fold+'_'+str(count)+'.wav', fs, r)
            count+=1
        print('Files saved!')
        
        time.sleep(2)
```
