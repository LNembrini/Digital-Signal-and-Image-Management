```python
def detection_face(directory1, directory2,size = (224, 224)):
    # Load model to detect frontal faces
    face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') 
    # variable for count errors
    f = 1
    err_noimg = 0
    
    for filename in os.listdir(directory1):
        # check for load only jpg and tiff images extension
        if filename.endswith(".jpg"):
             # load image          
            img = cv2.imread(os.path.join(directory1, filename))
            # conversion of image, this passage is necessary to pass images on face detector
            img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            faces = face_detector.detectMultiScale(img_gray)
            if len(faces) == 1:
                for (x, y, w, h) in faces:
                    # cut images on bases of segmentation of face detector
                    roi_color = img[y:y+h, x:x+w]
                    # resize of image
                    re_img = cv2.resize(roi_color, size)
                    cv2.imwrite(os.path.join(directory2, filename), re_img)
                    f += 1
                    
    
    # output          
    print('Created {}/{} images cropped'.format(f-1, f+err_noimg-1))
```

```python
net = VGGFace(include_top=False, weights='vggface', model='resnet50', pooling='max', input_shape=(512, 512, 3))
```

```python
# Dummy feature
def identity(image):
    return kimage.img_to_array(img).flatten()

# Data loader
def load_data(feature_extractor=identity):
    base_path = 'drive/MyDrive/DSIM retrieval/vip_volti_prova/'

    paths = []
    features = []

    for fi,f in enumerate(sorted(os.listdir(base_path))):
        if f.endswith('.jpg') and fi<maximages:
            # Save file path
            cur_path = base_path + f
            paths.append(cur_path)
            
            # Load file and extract features
            image = kimage.load_img(cur_path, target_size=(512, 512))
            cur_features = feature_extractor(image)
            features.append(cur_features)

    features = np.array(features)
    return features, paths
    #produce 2 output, per ogni img estraiamo features e teniamo traccia
    #del percorso file, così teniamo in memoria le descrizioni e non le img intere.
```

```python
def neural_features(img):
    x = kimage.img_to_array(img)
    x = preprocess_input(x, version=2)
    x = np.expand_dims(x, axis=0)
    f = net.predict(x)
    return f.flatten()
```

```python
maximages = 15788
```

```python
from time import time
```

```python
%time X_train, paths = load_data(feature_extractor=neural_features)
```
